```python
import nfldb

db = nfldb.connect()
```

```python
q = nfldb.Query(db)

q.game(season_year=2014, season_type='Regular')
players = [pp.player_id for pp in q.player(position='RB').sort('rushing_yds').limit(50).as_aggregate()]
```

```python
q = nfldb.Query(db)

q.game(season_year=2014)
q.play_player(player_id=players, rushing_att=1, fumbles_lost__ne=1)
carries = [ (p.player.full_name, p.rushing_yds) for p in q.as_play_players() ]
```

```python
c = {}
for name, yards in carries:
    if name not in c:
        c[name] = [yards]
    else:
        c[name].append(yards)
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

percentiles = [25, 50, 75]
names = []
m = []
percentile_m = { pct : list() for pct in percentiles }
att = []
total = []
for k, v in c.iteritems():
    m.append(np.mean(v))
    yd_percentiles = np.percentile(v, percentiles)
    for pct, yd_pct in zip(percentiles, yd_percentiles):
        percentile_m[pct].append(np.mean(filter(lambda x : x <= yd_pct, v)))
    att.append(len(v))
    total.append(sum(v))
    names.append(k)

report = pd.DataFrame(
    data={'NAMES' : names, 'YPC' : m}
)

for pct in percentiles:
    report['YPC_%s' % pct] = percentile_m[pct]

report['ATT'] = att
report['RSH_TOT'] = total

report.sort('YPC_75', ascending=False)
```
