```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import analysis

import nfldb

db = nfldb.connect()
```

```python
def graph_rush_vs_pass(passing, rushing, passing_pct, rushing_pct, title, threshold):
    """
    plot the rushing/passing percentages against score differential
    there are some very low occurrence score differentials e.g. -6 (which only happened twice)
    leave out the data for these low occurrence differentials
    """
    plt.figure(figsize=(20,5))
    
    p_diff = []
    p_pct = []
    for x, y in sorted(passing_pct.items()):
        if passing[x] + rushing[x] <= threshold:
            continue
        p_diff.append(x)
        p_pct.append(y)
     
    plt.plot(p_diff, p_pct, label='passing pct')
     
    r_diff = []
    r_pct = []
    for x, y in sorted(rushing_pct.items()):
        if passing[x] + rushing[x] <= threshold:
            continue
        r_diff.append(x)
        r_pct.append(y)
     
    plt.plot(r_diff, r_pct, label='rushing pct')
    plt.legend(loc=0)
    
    min_x = min( passing_pct.keys() + rushing_pct.keys() )
    max_x = max( passing_pct.keys() + rushing_pct.keys() )
    
    plt.xticks(np.arange(min_x, max_x + 1, 1.0))
    plt.yticks(np.arange(0, 101, 10.0))
    plt.title(title)
```

```python
q = nfldb.Query(db).game(season_year=2013, season_type='Regular')
passing, rushing, total, passing_pct, rushing_pct = analysis.rush_pass_breakdown(db, q, 'DEN')

total_passing_pct = np.mean(passing_pct.values())
print 'total passing pct: %0.2f%%\ntotal rushing pct: %0.2f%%' % (total_passing_pct, 100 - total_passing_pct)

plays_from_behind = sum(x[1] for x in filter(lambda x : x[0] < 0, total.items()))
plays_total = sum(x[1] for x in total.items())
print 'pct of plays from behind: %0.2f%%' % (plays_from_behind * 100. / plays_total)
```

```python
threshold = analysis.find_graph_threshold(total)
graph_rush_vs_pass(passing, rushing, passing_pct, rushing_pct, 'Denver 2013 rushing/passing percentages', threshold)
```
